```python
import pandas as pd

df = pd.read_csv("iris.csv")

x = df.drop(columns='Species')
y = df['Species']

from sklearn.tree import DecisionTreeClassifier, plot_tree
from matplotlib import pyplot as plt

irisModel = DecisionTreeClassifier()
irisModel.fit(x,y)

df[df['Petal.Length']<=2.45].shape

df[df['Petal.Length']<=2.45]['Species'].value_counts()

plt.figure(figsize=(14,20))
plot_tree(irisModel,feature_names=x.columns, filled=True, rounded=True,class_names=["Setosa","Versicolor","Virginica"])
plt.show()

```

# Hyperparameter Tuning

```python
iris_parameters = {'max_depth':[6,8,10],'min_samples_split':[4,6,7,9],'min_samples_leaf':[6,7,9]}

from sklearn.model_selection import GridSearchCV
gridModel = GridSearchCV(DecisionTreeClassifier(),param_grid=iris_parameters,cv=2)
gridModel.fit(x,y)
gridModel.best_estimator_
```
